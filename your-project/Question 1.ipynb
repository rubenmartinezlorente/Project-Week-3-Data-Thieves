### 1. What kind of categories of beers exists?

First let's read the csv of beers and categories

```python
import pandas as pd

beers = pd.read_csv('NEW_DATA/beers_new.csv')

categories = pd.read_csv('NEW_DATA/categories_new.csv')

breweries = pd.read_csv('NEW_DATA/brewery_new.csv')
```

```python
beers = beers.astype({'cat_id': int, 'brewery_id': int})
```

```python
beers.head()
```

```python
breweries.head()
```

```python
categories
```

```python
dataset1 = beers.merge(breweries, left_on='brewery_id', right_on ='id')
```

```python
dataset2 = dataset1.merge(categories, left_on='cat_id', right_on ='id')
```

```python
dataset2 = dataset2.drop(['id_x', 'id_y', 'address1', 'state'], axis=1)
```

```python
dataset2.head()
```

```python
dataset2.astype({'style_id': int})
```

```python
dataset2 = dataset2.rename(columns={"name_x": "beer_name", "name_y": "brewery_name"})
dataset2.head()
```

```python
dataset2['country'].value_counts().head()
```

```python
cat_popularity = dataset2['cat_name'].value_counts()
cat_popularity.head(8)
```

```python
categories = dataset2.groupby(['cat_name']).count()
```

```python
categories
```

```python
beer_per_category = categories.sort_values([('beer_name')], ascending=False)
beer_per_category
```

```python
beer_x_cat = dataset2.loc[:, ['cat_name','beer_name']]
beer_x_cat
```

```python
USA = dataset2.loc[dataset2['country'] == 'United States']
```

```python
USA_pop_beer = USA.groupby(['cat_name']).count().sort_values(by='beer_name', ascending=False)
USA_pop_beer.head(3)
```

```python
Belgium = dataset2.loc[dataset2['country'] == 'Belgium']
Belgium_pop_beer = Belgium.groupby(['cat_name']).count().sort_values(by='beer_name', ascending=False)
Belgium_pop_beer.head(3)
```

```python
Germany = dataset2.loc[dataset2['country'] == 'Germany']
Germany_pop_beer = Germany.groupby(['cat_name']).count().sort_values(by='beer_name', ascending=False)
Germany_pop_beer.head(3)
```

```python
Ireland = dataset2.loc[dataset2['country'] == 'Ireland']
Ireland_pop_beer = Ireland.groupby(['cat_name']).count().sort_values(by='beer_name', ascending=False)
Ireland_pop_beer.head(3)
```

```python
Netherlands = dataset2.loc[dataset2['country'] == 'Netherlands']
Netherlands_pop_beer = Netherlands.groupby(['cat_name']).count().sort_values(by='beer_name', ascending=False)
Netherlands_pop_beer.head(3)
```
